* üîπ range() function kya hota hai?

* range() Python ka built-in function hai jo numbers ki sequence (series) banane ke kaam aata hai.
* Ye zyada tar loops (for loop) ke saath use hota hai.
* üîπ range() kya return karta hai?
* The range() function returns a range object that generates a sequence of numbers.

* üëâ range() function ek range object return karta hai, list return nahi karta.

```python
r = range(5)
print(r)
```

```python
range(0, 5)
```

```python
# ‚ö†Ô∏è Agar numbers dekhne ho to:
print(list(range(5)))
```

* üîπ range() ke parameters me kya kya jata hai?
* range() me 3 tarah se parameters ja sakte hain üëá

```python
# 1Ô∏è‚É£ range(stop)
# üîπ Parameter:
# stop ‚Üí last number (exclude hota hai)
# Example:

range(5)
```

```python

# üëâ Sequence: 0, 1, 2, 3, 4
print(list(range(5)))
```

* 2Ô∏è‚É£ range(start, stop)
* üîπ Parameters:
* start ‚Üí kahan se start
* stop ‚Üí kahan tak (exclude)
* Example:

```python
range(2, 7)
```

```python
# üëâ Sequence: 2, 3, 4, 5, 6
print(list(range(2, 7)))
```

* 3Ô∏è‚É£ range(start, stop, step)
* üîπ Parameters:
* start ‚Üí starting number
* stop ‚Üí ending number (exclude)
* step ‚Üí kitne ka gap
*Example:

```python
range(1, 10, 2)
```

```python
# üëâ Sequence: 1, 3, 5, 7, 9
print(list(range(1, 10, 2)))
```

```python
# üîπ Negative step example
print(list(range(10, 0, -2)))
```

* üîπ sum() function kya hota hai?
* sum() Python ka built-in function hai jo numbers ka total (addition) nikalne ke liye use hota hai.
üîπ sum() kya return karta hai?
* The sum() function returns the total of all numeric elements in an iterable.
* üëâ sum() function ek number return karta hai
* Agar numbers integer hain ‚Üí return int
* Agar numbers me float hai ‚Üí return float
* Example:

```python
sum([1, 2, 3])
```

```python
sum([1.5, 2.5])
```

* üîπ sum() ke parameters me kya kya jata hai?
* sum() me 2 parameters ho sakte hain üëá
* sum(iterable, start)

* 1Ô∏è‚É£ First parameter: iterable (mandatory)

* üëâ Ye wo collection hoti hai jisme numbers hote hain:

* Allowed iterables:

* List
* Tuple
* Set
* Range
* Examples:

```python
# üîπ List
sum([10, 20, 30])
```

```python
# üîπ Tuple
sum((1, 2, 3))
```

```python
# üîπ Set
sum({4, 5, 6})
```

```python

# üîπ Range
sum(range(1, 6))
# üëâ 1 + 2 + 3 + 4 + 5 = 15
```

* 2Ô∏è‚É£ Second parameter: start (optional)
* üëâ Starting value jisse addition start hota hai
* Example:

```python
sum([1, 2, 3], 10)
```

* # lambda function
* Lambda function Python ka short & anonymous function hota hai.
* Anonymous ka matlab üëâ iska koi naam nahi hota (ya short name me use hota hai).
* A lambda function is a one-line function that is used for small, simple operations.
* üëâ It is mainly used when:
* Function ka logic bahut chhota ho
* Function ko sirf ek hi jagah use karna ho
* ‚úÖ Syntax of Lambda 
* lambda arguments : expression
* ‚ö†Ô∏è Important points:

* lambda keyword use hota hai

* Arguments colon (:) se pehle aate hain

* Expression colon ke baad hota hai

* Lambda function sirf ek expression likh sakta hai

* return keyword likhne ki zarurat nahi hoti
* example

```python
square = lambda x: x * x
print(square(6))
```

* same example using normal function

```python
def square(x):
    return x * x

print(square(6))
```
